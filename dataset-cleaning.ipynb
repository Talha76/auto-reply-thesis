```python
with open("./datasets/contents.csv") as f:
    reviews = f.readlines()
len(reviews)
```

```python
clean_reviews = [clean_text(review) for review in reviews if clean_text(review)]
len(clean_reviews)
```

```python
clean_reviews
```

```python
with open("datasets/clean_contents.csv", "w") as f:
    for review in clean_reviews:
        f.write(review + "\n")
```
